```python
# Get number density
from scipy.integrate import simps

def integ(x,y):
    Ml = 7.0
    Y7 = interp(Ml, x, y)
    ids=x>Ml
    x=x[ids]
    y=y[ids]
    x=np.append(x,Ml)
    y=np.append(y,Y7)
    ids=np.argsort(x)
    x=x[ids]
    y=y[ids]
    return simps(10.**y,x), simps(10.**y*10**x,x)
```

```python
# intergration with another method
from scipy.integrate import simps,fixed_quad,quad
from scipy.interpolate import CubicSpline
def inteq(x,y):
    ids=np.argsort(x)
    x=x[ids]
    y=y[ids]
    cs=CubicSpline(x, y)
    def innerfunc(vec):
        return 10.**cs(vec)

    def innerfunc2(vec2):
        return 10.**cs(vec2)*10.**vec2
    return quad(innerfunc,7.0,10), quad(innerfunc2,7.0,10)
```
